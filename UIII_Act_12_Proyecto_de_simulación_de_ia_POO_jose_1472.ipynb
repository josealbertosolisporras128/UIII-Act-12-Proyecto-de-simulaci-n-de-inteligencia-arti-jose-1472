```python
class Maquillista:
    def __init__(self):
        # Línea 1: El constructor crea la lista interna 'plantilla' inicialmente vacía [4, 5].
        self.plantilla = []

    def agregar_maquillista(self):
        # Línea 2: Captura interactiva de los 7 atributos requeridos para el objeto [6, 7].
        print("\n--- Registro de Nueva Maquillista ---")
        nombre = input("Nombre de la artista: ")
        kit = input("Kit de marcas: ")
        especialidad = input("Especialidad (Social/FX): ")
        estudio = input("Estudio de maquillaje (Si/No): ")
        redes = input("Redes sociales: ")
        agenda = input("Agenda (Horarios): ")
        certificaciones = input("Certificaciones: ")

        # Línea 3: Agrupa los atributos en un diccionario para representar al objeto [1, 8].
        nuevo_perfil = {
            "nombre": nombre, "kit": kit, "especialidad": especialidad,
            "estudio": estudio, "redes": redes, "agenda": agenda,
            "certif": certifications if 'certifications' in locals() else certificaciones
        }

        # Línea 4: Añade el nuevo objeto al final de la colección mediante append() [3, 9].
        self.plantilla.append(nuevo_perfil)
        print(f"IA: {nombre} ha sido guardada en el sistema.")

    def listar_maquillista(self):
        # Línea 5: Verifica si existen registros antes de intentar mostrarlos [10].
        if not self.plantilla:
            print("\nIA: No hay registros para mostrar.")
        else:
            # Línea 6: Distribuye la información en una tabla bien alineada usando f-strings [11-13].
            print("\n" + "="*140)
            print(f"{'ID':<4} | {'Nombre':<15} | {'Especialidad':<12} | {'Kit':<15} | {'Estudio':<10} | {'Redes':<15} | {'Agenda':<15} | {'Certificaciones'}")
            print("-" * 140)
            for i, m in enumerate(self.plantilla):
                print(f"{i:<4} | {m['nombre']:<15} | {m['especialidad']:<12} | {m['kit']:<15} | {m['estudio']:<10} | {m['redes']:<15} | {m['agenda']:<15} | {m['certif']}")
            print("="*140)

    def actualizar_maquillista(self):
        # Línea 7: Reemplaza una maquillista por otra completamente nueva en la misma posición [1, 2].
        self.listar_maquillista()
        if self.plantilla:
            indice = int(input("\nIngrese el ID de la maquillista que desea CAMBIAR (reemplazar): "))
            if 0 <= indice < len(self.plantilla):
                print(f"\n--- Ingrese los datos de la NUEVA artista para el ID {indice} ---")
                # Se solicitan todos los datos de nuevo para el intercambio total [14].
                nuevo_nombre = input("Nuevo Nombre: ")
                nuevo_kit = input("Nuevo Kit: ")
                nueva_espec = input("Nueva Especialidad: ")
                nuevo_estud = input("Nuevo Estudio: ")
                nuevas_redes = input("Nuevas Redes: ")
                nueva_agenda = input("Nueva Agenda: ")
                nuevas_cert = input("Nuevas Certificaciones: ")

                # Línea 8: Se sobrescribe la posición del índice con el nuevo diccionario [1, 2].
                self.plantilla[indice] = {
                    "nombre": nuevo_nombre, "kit": nuevo_kit, "especialidad": nueva_espec,
                    "estudio": nuevo_estud, "redes": nuevas_redes, "agenda": nueva_agenda,
                    "certif": nuevas_cert
                }
                print(f"IA: Registro {indice} actualizado. Ahora es {nuevo_nombre}.")
            else:
                print("IA Error: ID fuera de rango.")

    def borrar_maquillista(self):
        # Línea 9: Elimina físicamente el registro de la lista usando su ID [9, 10, 15].
        self.listar_maquillista()
        if self.plantilla:
            indice = int(input("\nID de la maquillista a borrar: "))
            if 0 <= indice < len(self.plantilla):
                eliminado = self.plantilla.pop(indice)
                print(f"IA: {eliminado['nombre']} ha sido eliminada de la plantilla.")
            else:
                print("IA Error: ID inválido.")

# --- Menú Principal ---
def menu():
    sistema = Maquillista() # Instanciación de la clase [3, 16]
    while True:
        print("\n********** SISTEMA IA BEAUTY-OO 2026 **********")
        print("1. Agregar Maquillista")
        print("2. Listar Maquillistas (Tabla)")
        print("3. Cambiar Maquillista (Actualización Total)")
        print("4. Borrar Maquillista")
        print("5. Salir")
        opcion = input("Seleccione opción: ")

        if opcion == "1":
            sistema.agregar_maquillista()
        elif opcion == "2":
            sistema.listar_maquillista()
        elif opcion == "3":
            sistema.actualizar_maquillista()
        elif opcion == "4":
            sistema.borrar_maquillista()
        elif opcion == "5":
            print("Cerrando sesión inteligente...")
            break

if __name__ == "__main__":
    menu()
```
